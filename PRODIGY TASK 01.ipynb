## Dropping unnecessary columns

```python
df.drop(['Series Name', 'Country Code'], axis=1, inplace= True)
df.columns
```

## Extraction of Top 10 Countries 

```python
# Filter data for Total Population
total_population= df[df['Series Code'] == 'SP.POP.TOTL']

# Sorting the data based on 2022 population
total_population_sorted= total_population.sort_values(by= "2022", ascending= False)

# Getting the Top 10 Countries with highest 2022 population
top_ten_countries= total_population_sorted.head(10)
print("Top 10 Countries of Total Population\n")
print(top_ten_countries[['Country Name']])
```

## Top 10 Countries of Total Population in the Year 2022 and 2016

```python
# Creating Bar plot
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x= "2022", y= "Country Name", data= top_ten_countries, palette= "coolwarm")
plt.title("Top 10 Countries of Total Population 2022", fontsize= 10)
plt.xlabel("Total Population", fontsize= 10)
plt.ylabel("Country", fontsize= 10)
plt.show()

plt.figure(figsize= (15,6))
plt.subplot(2,2,2)
sns.barplot(x= "2016", y= "Country Name", data= top_ten_countries, palette= "coolwarm")
plt.title("Top 10 Countries of Total Population 2016", fontsize= 10)
plt.xlabel("Total Population", fontsize= 10)
plt.ylabel("Country", fontsize= 10)
plt.show()
```

## Top 10 Countries of Total Population in the Year 2010 and 2001 

```python
# Creating Bar plot
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x= "2010", y= "Country Name", data= top_ten_countries, palette= "coolwarm")
plt.title("Top 10 Countries of Total Population 2010", fontsize= 10)
plt.xlabel("Total Population", fontsize= 10)
plt.ylabel("Country", fontsize= 10)
plt.show()

plt.figure(figsize= (15,6))
plt.subplot(2,2,2)
sns.barplot(x= "2001", y= "Country Name", data= top_ten_countries, palette= "coolwarm")
plt.title("Top 10 Countries of Total Population 2001", fontsize= 10)
plt.xlabel("Total Population", fontsize= 10)
plt.ylabel("Country", fontsize= 10)
plt.show()
```

## STACKED BARPLOT

## Top 10 Countries with Male and Female Population 2022

```python
# Merging male and female population
merged_data= pd.merge(male_population, female_population, on= "Country Name", suffixes= ("_male", "_female"))
```

```python
merged_data
```

```python
# Calculating total population for each country
merged_data["Total Population"]= merged_data["2022_male"] + merged_data["2022_female"]
```

```python
merged_data.head()
```

```python
# Sorting data in descending order
sorted_data= merged_data.sort_values(by= "Total Population", ascending= False)

# Selecting top 10 countries
top_10_countries= sorted_data.head(10)
```

```python
# Creating Stacked bar plot
plt.figure(figsize=(12,6))
sns.barplot(x= "Country Name", y= "2022_female", data= top_10_countries, color= "red", label= "Female Population")
sns.barplot(x= "Country Name", y= "2022_male", data= top_10_countries, bottom= top_10_countries["2022_female"], color= "blue", label= "Male Population")
plt.title("Top 10 Countries with Male & Female Population 2022")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha= "right")
plt.show()
```

## Bottom 10 Countries

```python
# Selecting bottom 10 Countries
bottom_10_countries= sorted_data.tail(10)
```

```python
# Creating Stacked bar plot
plt.figure(figsize=(12,6))
sns.barplot(x= "Country Name", y= "2022_female", data= bottom_10_countries, color= "red", label= "Female Population")
sns.barplot(x= "Country Name", y= "2022_male", data= bottom_10_countries, bottom= bottom_10_countries["2022_female"], color= "blue", label= "Male Population")
plt.title("Bottom 10 Countries with Male & Female Population 2022")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha= "right")
plt.show()
```
